```python
import os
import cv2
import tqdm
import numpy as np
import pandas as pd
```

```python
def build_df(template, cell):
    for i in range(len(cell)):
        template[f'Px_{i:03d}'] = cell[i]
    return template

def transform(path = None):
    df = pd.DataFrame()
    list_images = os.listdir(path)
    for idx in tqdm.tqdm(range(len(list_images))):
        k = 0
        img = cv2.imread(f'{path}/{list_images[idx]}', 0)
        for i in range(0, 250, 28):
            for j in range(0, 250, 28):
                cell = img[i:i+28, j:j+28]
                cell = cell.flatten()
                template = {'Label': list_images[idx][k]}
                df = df.append(build_df(template, cell), ignore_index = True)
                k += 1
    return df[['Label'] + [f'Px_{i:03d}' for i in range(784)]]

pixels_df = transform(path = '/content/drive/MyDrive/Projects/Sudoku/PrcPix')
pixels_df = pixels_df.sort_values(['Label'], ascending = [True])
pixels_df.to_excel('/content/drive/MyDrive/Projects/Sudoku/data003.xlsx', index = False)
```
